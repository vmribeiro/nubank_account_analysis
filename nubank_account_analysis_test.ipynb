##### 2. Tratamento e criação do dataframe
<br>
<br>
Essa parte do código trata os dados do arquivo .txt criado após a extração, e cria um dataframe pronto para uso em outras bibliotecas de acordo com os dados existentes no extrato de suas transações.
<br><br>


```python
import json
from pynubank import Nubank, MockHttpClient
import pandas as pd

def get_df_card_statements_from_txt(filename = r'/Users/viribeiro/Downloads/Personal/Github/files/extrato_nubank.txt'):
    with open(filename,"r") as f:
        card_statements = f.read()
    card_statements = json.loads(str(card_statements).replace("'", '"').replace('False', 'false').replace('True', 'true'))
    # transform results into a dataframe
    df_temp = None
    df_card_statements = pd.DataFrame({
          'description': ['dummy'],                               
          'category': ['dummy'],     
          'amount': ['dummy'],     
          'time': ['dummy'],     
          'source': ['dummy'],     
          'title': ['dummy'],     
          'amount_without_iof': ['dummy'],     
          'account': ['dummy'],     
          'status': ['dummy'],     
          'num_installments': ['dummy'],     
          'total_installments_amount': ['dummy'],     
          'id': ['dummy'],     
          '_links': ['dummy'],     
          'tokenized': ['dummy'],     
          'href': ['dummy']
    })

    for i in range(0,len(card_statements)):
        try:
            num_installments = str(card_statements[i].get('details')['charges']['count']).replace("'",'').replace('[','').replace(']','')       
            total_installments_amount = str(card_statements[i].get('details')['charges']['amount']).replace("'",'').replace('[','').replace(']','')    
        except:
            num_installments= 0
            total_installments_amount = 0
            pass 
        try: 
            status = str(card_statements[i].get('details')['status']).replace("'",'').replace('[','').replace(']','')
        except:
            status = 'not defined'
            pass
        df_temp = pd.DataFrame(data = {   'description': [str(card_statements[i].get('description')).replace("'",'').replace('[','').replace(']','')],                               
                                          'category': [str(card_statements[i].get('category')).replace("'",'').replace('[','').replace(']','')],
                                          'amount': [str(card_statements[i].get('amount')).replace("'",'').replace('[','').replace(']','')],
                                          'time': [str(card_statements[i].get('time')).replace("'",'').replace('[','').replace(']','')],
                                          'source': [str(card_statements[i].get('source')).replace("'",'').replace('[','').replace(']','')],
                                          'title': [str(card_statements[i].get('title')).replace("'",'').replace('[','').replace(']','')],
                                          'amount_without_iof': [str(card_statements[i].get('amount_without_iof')).replace("'",'').replace('[','').replace(']','')],
                                          'account': [str(card_statements[i].get('account')).replace("'",'').replace('[','').replace(']','')],
                                          'status': [status],                                       
                                          'num_installments': [num_installments],
                                          'total_installments_amount': [total_installments_amount],
                                          'id': [str(card_statements[i].get('id')).replace("'",'').replace('[','').replace(']','')],
                                          '_links': [str(card_statements[i].get('_links')).replace("'",'').replace('[','').replace(']','')],
                                          'tokenized': [str(card_statements[i].get('tokenized')).replace("'",'').replace('[','').replace(']','')],
                                          'href': [str(card_statements[i].get('href')).replace("'",'').replace('[','').replace(']','')]
                                        })
        df_card_statements = pd.concat([df_card_statements, df_temp])
    df_card_statements = df_card_statements[df_card_statements['description'] != 'dummy']
        
    return df_card_statements
```

```python
df_card_statements = get_df_card_statements_from_txt(filename = r'/Users/viribeiro/Downloads/Personal/Github/files/extrato_nubank.txt')
df_card_statements.head(5) 
```

```python
from datetime import datetime

df_card_statements = df_card_statements.fillna('nan')
df_card_statements['description'] = df_card_statements['description'].astype(str)
df_card_statements['category'] = df_card_statements['category'].astype(str)
df_card_statements['source'] = df_card_statements['source'].astype(str)
df_card_statements['title'] = df_card_statements['title'].astype(str)
df_card_statements['title'] = df_card_statements.apply(lambda row: str(row['title']).capitalize(), axis = 1)
df_card_statements['status'] = df_card_statements['status'].astype(str)
df_card_statements['id'] = df_card_statements['id'].astype(str)
df_card_statements['_links'] = df_card_statements['_links'].astype(str)
df_card_statements['tokenized'] = df_card_statements['tokenized'].astype(str)
df_card_statements['href'] = df_card_statements['href'].astype(str)
df_card_statements['transaction_date'] =  df_card_statements.apply(lambda row: datetime.strptime(str(row['time']).split('T')[0], '%Y-%m-%d'), axis = 1)    
df_card_statements['amount'] =  df_card_statements.apply(lambda row: float(0) if str(row['amount']).replace(' ', '') == 'nan' else float(str(row['amount']).replace(' ', '')), axis = 1)   
df_card_statements['amount_without_iof'] =  df_card_statements.apply(lambda row: float(0) if str(row['amount_without_iof']).replace(' ', '') == 'nan' or str(row['amount_without_iof']).replace(' ', '') == 'None' else float(str(row['amount_without_iof']).replace(' ', '')), axis = 1)    
df_card_statements['num_installments'] =  df_card_statements.apply(lambda row: float(0) if str(row['num_installments']).replace(' ', '') == 'nan' else float(str(row['num_installments']).replace(' ', '')), axis = 1)    
df_card_statements['total_installments_amount'] =  df_card_statements.apply(lambda row: float(0) if str(row['total_installments_amount']).replace(' ', '') == 'nan' else float(str(row['total_installments_amount']).replace(' ', '')), axis = 1)    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.float_format = '{:.0f}'.format

# group by all values
df_amount_by_title = df_card_statements.groupby('title').sum().reset_index().filter(items = ['title', 'amount']).sort_values(by = 'amount', ascending = False)              

sns.set_theme(style="white")
# creating figure
f, ax = plt.subplots()
ax.ticklabel_format(style='plain', axis='both')
ax = sns.barplot(x="amount", y="title", data=df_amount_by_title)
sns.despine()
```
